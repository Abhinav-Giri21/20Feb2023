```python
'''Q1.Explain GET and POST methods.
Ans.GET:It is used to request data from a specified resource. When a client sends a GET request, the server retrieves the requested data and sends it 
back as a response. GET requests can be cached, bookmarked, and are visible in the URL bar of a web browser. They are generally used for retrieving 
data.

POST:It is used to submit data to be processed to a specified resource. When a client sends a POST request, the data is sent in the request body, 
separate from the URL. The server processes the data and sends back a response. POST requests are not cached, cannot be bookmarked, and do not show 
the submitted data in the URL. They are commonly used for creating or updating data on the server.'''
```

```python
'''Q2.Why is request used in Flask?
Ans.In Flask, the request object is used to access incoming request data. It allows you to retrieve information such as form data, URL parameters,
headers, cookies, and more from the client's request.'''
```

```python
'''Q3.Why is redirect() used in Flask?
Ans.The redirect() function in Flask is used to redirect the client to a different route or URL. It is commonly used when you want to redirect the 
user after performing a certain action, such as after submitting a form or completing a login process.'''
```

```python
'''Q4.What are templates in Flask? Why is the render_template() function used?
Ans.Templates in Flask are files that contain HTML and other dynamic elements. They are used to generate dynamic web pages by combining the HTML 
structure with data from the application.

The render_template() function is used to render these templates in Flask.'''
```
